```python
import requests
import datetime
from IPython.display import clear_output

where = input('Whrere or what you want to get news from')
where = where.replace(' ','%20')
clear_output()

year = datetime.datetime.today().year
month = datetime.datetime.today().month
day = datetime.datetime.today().day

req = requests.get(f'https://newsapi.org/v2/everything?qInTitle={where}&from={year}-{month}-{day}&to={year}-{month}-{day-1}&sortBy=popularity&language=en&apiKey=890603a55bfa47048e4490069ebee18c')

content = req.json()


for item in content['articles']:
    print('\n')
    print(item['title'])
    print(item['description'])
    
```
